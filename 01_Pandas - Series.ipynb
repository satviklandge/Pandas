## Pandas Series
    Pandas Series is a One Dimensional indexed array. It is most similar to the NumPy array that can store data of any type (integers, floats, strings, Python objects, etc.).

### Import Pandas

```python
import pandas as pd
```

### Check Version 

```python
pd.__version__
```

### Creating Pandas Series

```python
ser_1 = pd.Series([1, 4, 6, -8, 6.9, "Pandas", False])
ser_1
```

```python
# Check the type of ser_1 object
type(ser_1)
```

### Creating Empty Series

```python
ser_2 = pd.Series([])
ser_2
```

### Creating Series Using a different method 

```python
# Changing the index of series
ser_3 = pd.Series([1, 2, 3, 4, 5, 6], index = ['a', 'b', 'c', 'd', 'e', 'f'])
ser_3
```

```python
# Changing data Type of series into float
ser_4 = pd.Series([23, 45, 65, 89, 39], index = ['a', 'b', 'c', 'd', 'e'], dtype = float)
ser_4
```

### Creating series from scalar values ( scalar values means single value)

```python
ser1_scalar = pd.Series(4,index = [1,2,3,4,5,6])
ser1_scalar
```

### Creating Series From Python dictionary 

```python
ser_dict = pd.Series({"Sr_no" : [1,2,3,4,5], "Name": ['Satvik', 'Harsh', 'Sumit', 'Yash', 'Laksh'], "Roll_no" : [101,102,103,104,105]})
ser_dict
```

```python
# Accessing element from series
ser_dict[1]
```

```python
# Slicing Series
ser_dict[1:]
```

## Math Operation

```python
# Adding Two series
ser_5 = pd.Series([23, 67, 84, 76, 99])
ser_6 = pd.Series([47, 82, 61, 59, 31])

# Method 1
A = ser_5 + ser_6
print(A)

# Method 2 
ser_5.add(ser_6)
```

```python
# Find Min , Max , Smallest Element of Series
print("Mean of series s5 ::",ser_5.mean())
print("Max Element of series s5 ::",ser_5.max())
print("Samllest Element of series s5 ::",ser_5.min())
```
