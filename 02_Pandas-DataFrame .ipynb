## Pandas Dataframe with different ways
    A Pandas DataFrame is a two-dimensional, labeled data structure in Python, similar to a table in Excel or a SQL database.DataFrames are widely used in data analysis, data cleaning, and machine learning because they make it easy to:

    - Store structured data
    
    - Filter and sort records
    
    - Handle missing values
    
    - Perform statistical operations
    
    - Merge and join datasets
    

### import pandas

```python
import pandas as pd
```

### Creating empty dataframe

```python
emt_df = pd.DataFrame()
print(emt_df)
```

### Creating Dataframe With Single Column

```python
pd_df1 = pd.DataFrame(['A', 'B', 'C', 'D'])
print(pd_df1)
```

### Creating Dataframe With Multiple Column

```python
pd_df2 = pd.DataFrame([[1, 2, 3], [4, 5, 6], [7, 8, 9]])
print(pd_df2)
```

### Creating Dataframe By Dictionary 

```python
pd_dict = {"Sr_no" : [1,2,3,4,5], "Name": ['Satvik', 'Harsh', 'Sumit', 'Yash', 'Laksh'], "Roll_no" : [101,102,103,104,105]}
pd_df3 = pd.DataFrame(pd_dict)
pd_df3
```
